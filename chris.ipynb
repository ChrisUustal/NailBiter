```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
data = np.loadtxt('Datasets/simple_dataset.csv',delimiter = ',')
x = data[:,0:-1]
y = data[:,-1:] 
num_samples = np.shape(data)[0]

print("Your dataset has", num_samples, "elements with", np.shape(x)[1], "features")
```

```python
for i in range(np.shape(x)[1]):
    plt.plot(x[:,i], label=i)
plt.legend()
plt.show()
```

```python
model = LogisticRegression().fit(x, y[:,0])
print(np.size(y) - np.sum(model.predict(x) == y[:,0]))
```
